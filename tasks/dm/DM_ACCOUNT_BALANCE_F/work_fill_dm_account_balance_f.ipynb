```python
from fill_dm_account_balance_f import fill_dm_account_balance_f
from pyspark.sql import SparkSession, Row
from pyspark.sql import functions as F
from datetime import datetime
import time
import uuid
```

```python
spark = SparkSession.builder \
    .appName("fill_dm_account_balance") \
    .enableHiveSupport() \
    .getOrCreate()

spark.sql('Use DM')
```

```python
dates_jan_2018 = [f"2018-01-{day:02d}" for day in range(1, 32)]
print(dates_jan_2018)
```

```python
start_time = datetime.now()
job_name = "fill_dm_account_balance_f"
status = "SUCCESS"
for date_str in dates_jan_2018:
    df = fill_dm_account_balance_f(date_str, spark)
    print(f"{date_str}: строк в df = {df.count()}")
    df.show(5)
    df.write.mode("append").insertInto("dm.dm_account_balance_f")
time.sleep(5)

end_time = datetime.now()

log_row = Row(
    id=str(uuid.uuid4()),
    job_name=job_name,
    start_time=start_time,
    end_time=end_time,
    status=status
)
log_df = spark.createDataFrame([log_row])
log_df.write.mode("append").insertInto("logs.LOGS")
```
