```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql import DataFrame
from datetime import datetime, timedelta
spark = SparkSession.builder \
    .appName("fill_account_balance_f") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
on_date = '2018-01-01'#форматы
```

```python
md_account_d = spark.read.parquet("/user/hive/warehouse/ds.db/md_account_d")
ft_balance_f = spark.read.parquet('/user/hive/warehouse/ds.db/ft_balance_f')
dm_account_balance_f = spark.read.parquet('/user/hive/warehouse/dm.db/dm_account_balance_f')
dm_account_turnover_f = spark.read.parquet(f'/user/hive/warehouse/dm.db/dm_account_turnover_f/')
dm_account_turnover_f.show(100)
```

```python
md_account_d.show()
```

```python
md_exchange_rate_d = spark.read.parquet('/user/hive/warehouse/ds.db/md_exchange_rate_d')
md_exchange_rate_d.show()
```

```python
ruble_exchange_row = md_exchange_rate_d \
    .filter(F.col("CODE_ISO_NUM") == 643) \
    .select("REDUCED_COURCE") \
    .limit(1) \
    .collect()

if ruble_exchange_row:
    ruble_exchange_rate = ruble_exchange_row[0]["REDUCED_COURCE"]
else:
    ruble_exchange_rate = 1.0
```

```python
# on_date = '2018-01-01'
on_date_obj = datetime.strptime(on_date, '%Y-%m-%d')
prev_date_str = (on_date_obj - timedelta(days=1)).strftime('%Y-%m-%d')

date_as_date = F.to_date(F.lit(on_date), "yyyy-MM-dd")
prev_date = F.date_sub(date_as_date, 1)
prev_date_formatted = F.date_format(prev_date, "dd.MM.yyyy")

dm_account_balance_f.printSchema()

dm_account_balance_f = dm_account_balance_f.where(F.col('on_date') == prev_date_str)\
.select('on_date', 'account_rk', 'balance_out', (F.col('balance_out') * ruble_exchange_rate).alias('balance_out_rub'))
```

```python
dm_account_balance_f.show()
```

```python
md_account_d_between_on_date = md_account_d.where(date_as_date.between(F.col("DATA_ACTUAL_DATE"), F.col("DATA_ACTUAL_END_DATE")))
```

```python
md_account_d_between_on_date.show()
```

```python
prev_balances = dm_account_balance_f.filter(
    F.col("on_date") == prev_date
).select("account_rk", "balance_out")
prev_balances.show()
```

```python
#обороты по счёту
turnovers = dm_account_turnover_f.filter(
    F.col("on_date") == on_date
).select("account_rk", "debet_amount", "credit_amount")
turnovers.show()
```

```python
# Добавим остатки (если нет — 0)
accounts = md_account_d_between_on_date \
    .join(prev_balances, on="account_rk", how="left") \
    .withColumn("balance_out", F.coalesce(F.col("balance_out"), F.lit(0.0)))
```

```python
# Добавим обороты (если нет — 0)
accounts = accounts.join(turnovers, on="account_rk", how="left") \
    .withColumn("debet_amount", F.coalesce(F.col("debet_amount"), F.lit(0.0))) \
    .withColumn("credit_amount", F.coalesce(F.col("credit_amount"), F.lit(0.0)))
accounts.show()
```

```python
accounts = accounts.withColumn(
    "balance_out_rub",
    F.when(F.col("char_type") == "А",
         F.col("balance_out") + F.col("debet_amount") - F.col("credit_amount")
    ).when(F.col("char_type") == "П",
         F.col("balance_out") - F.col("debet_amount") + F.col("credit_amount")
    ).otherwise(F.col("balance_out"))  # если тип неизвестен
)
```

```python
accounts.show()
```

```python
accounts = accounts.select('ACCOUNT_RK', 'balance_out', ('balance_out_rub')).withColumn('on_date', F.lit(date_as_date))
```

```python

    # Получение курса
    ruble_exchange_row = md_exchange_rate_d \
        .filter(F.col("CODE_ISO_NUM") == 643) \
        .select("REDUCED_COURCE") \
        .limit(1) \
        .collect()
    
    ruble_exchange_rate = ruble_exchange_row[0]["REDUCED_COURCE"] if ruble_exchange_row else 1.0

    # Получение предыдущей даты как строки
    prev_date_str = (datetime.strptime(on_date, "%Y-%m-%d") - timedelta(days=1)).strftime("%d.%m.%Y")

    # Фильтрация по строковым значениям
    dm_account_balance_f = dm_account_balance_f\
        .where(F.col('on_date') == prev_date_str)\
        .select('on_date', 'account_rk', 'balance_out', 
                (F.col('balance_out') * ruble_exchange_rate).alias('balance_out_rub'))

    # md_account_d: фильтрация по актуальности
    md_account_d_between_on_date = md_account_d\
        .where((F.lit(on_date) >= F.col("DATA_ACTUAL_DATE")) & (F.lit(on_date) <= F.col("DATA_ACTUAL_END_DATE")))

    # Остатки на предыдущую дату
    prev_balances = spark.table("dm.dm_account_balance_f") \
    .filter(F.col("on_date") == prev_date_str) \
    .select("account_rk", "balance_out", "balance_out_rub")

    # Обороты за текущую дату
    turnovers = dm_account_turnover_f.filter(F.col("on_date") == on_date)\
        .select("account_rk", "debet_amount", "credit_amount")

    # Счета + остатки
    accounts = md_account_d_between_on_date \
        .join(prev_balances, on="account_rk", how="left") \
        .withColumn("balance_out", F.coalesce(F.col("balance_out"), F.lit(0.0)))

    # Добавляем обороты
    accounts = accounts.join(turnovers, on="account_rk", how="left") \
        .withColumn("debet_amount", F.coalesce(F.col("debet_amount"), F.lit(0.0))) \
        .withColumn("credit_amount", F.coalesce(F.col("credit_amount"), F.lit(0.0)))
    
    # Расчёт рублёвого остатка
    accounts = accounts.withColumn(
        "balance_out_rub",
        F.when(F.col("char_type") == "А",
             F.col("balance_out") + F.col("debet_amount") - F.col("credit_amount")
        ).when(F.col("char_type") == "П",
             F.col("balance_out") - F.col("debet_amount") + F.col("credit_amount")
        ).otherwise(F.col("balance_out"))
    )

    # Возврат с правильной on_date
    accounts = accounts.select('account_rk', 'balance_out', 'balance_out_rub')\
        .withColumn('on_date', F.lit(on_date))
print(f"on_date = {on_date}, prev_date = {prev_date_str}")
accounts.show()
```
