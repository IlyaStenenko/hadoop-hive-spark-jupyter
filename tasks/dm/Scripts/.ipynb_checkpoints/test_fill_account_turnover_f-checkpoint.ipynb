```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
spark = SparkSession.builder \
    .appName("fill_account_turnover_f") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
ft_posting_f = spark.read.parquet('/user/hive/warehouse/ds.db/ft_posting_f')
```

```python
ft_posting_f.show()
```

```python
posting_on_date = ft_posting_f.where(F.col('oper_date') == '09-01-2018')
credit_account_rk = posting_on_date.select(F.col('credit_account_rk').alias('account_rk'))
debet_account_rk = posting_on_date.select(F.col('debet_account_rk').alias('account_rk'))
account_rk = credit_account_rk.union(debet_account_rk).distinct();
account_rk.show()
account_rk.agg(F.count('account_rk')).show()
```

```python
#posting_on_date = posting_on_date.join(account_rk, posting_on_date['credit_account_rk'] == account_rk['account_rk'], how = 'inner') # Нужно ли?
posting_on_date.agg(F.count('account_rk')).show()
posting_on_date.show()
credit_amount = posting_on_date.groupby('credit_account_rk')\
.agg(F.sum('credit_amount')).select
credit_amount.show()
```
